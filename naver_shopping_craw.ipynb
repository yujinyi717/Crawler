```python
import re, requests, csv 
from bs4 import BeautifulSoup 
from time import sleep
import pandas as pd

# url 가져오기 
url = "https://search.shopping.naver.com/catalog/29938164618?query=lg%20tv&NaPm=ct%3Dl1w5iim8%7Cci%3Dc809d139f2ae38d6cbb5fb2c91d2965c9ec4fbf7%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Dbaa6d0cf44172f61736be61218f8cbfb1baf16fa"
soup = BeautifulSoup(requests.get(url).text, 'html.parser') 
soup

# 제품명
elements = soup.select('title')[0].text
elements

# 리뷰 타이틀 
title = soup.findAll("em", {"class": "reviewItems_title__39Z8H"})
title = [i.get_text() for i in title]

# 리뷰 컨텐츠 
content = soup.findAll("p", {"class": "reviewItems_text__XIsTc"})
content = [i.get_text() for i in content]

# 리뷰 스코어
score = soup.findAll("span", {"class": 'reviewItems_average__16Ya-'})
score = [i.text[-1:] for i in score]

# 구매처, 닉네임, 작성일자
mall_name_date = soup.findAll("span", {"class": "reviewItems_etc__1YqVF"})
mall_name_date = [i.get_text() for i in mall_name_date]

# 구매처
mall = mall_name_date[::3]

# 닉네임
name = mall_name_date[1::3]

# 작성일자
date = mall_name_date[2::3]

# 데이터 프레임 생성 
df = pd.DataFrame({'title':title,'content':content,'score':score,'mall':mall,'name':name,'date':date })
df
```
