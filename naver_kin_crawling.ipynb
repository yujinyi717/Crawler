```python
import time
import selenium
import pandas as pd
from selenium import webdriver
import pandas as pd
```

# 1. 지식인 url 추출 

```python
# 블로그 제목, url, 요약문, 날짜 

def get_post_info(keyword: str, start_date: str, end_date: str, pages: int) -> pd.DataFrame:
    total_header = []
    total_url = []
    total_contents = []
    total_date = []
    # chromedriver path
    driver = webdriver.Chrome("D://chromedriver.exe")
    
    for j in range(1, pages + 1) :
        # first page
        first_url = f'https://kin.naver.com/search/list.nhn?sort=none&section=kin&query={keyword}&period={start_date}%7C{end_date}.'
        # After first page
        loop_url = f"https://kin.naver.com/search/list.nhn?sort=none&query={keyword}&period={start_date}%7C{end_date}&section=kin&page={j}"
        driver.get(first_url)

        if j != 1:
            driver.get(loop_url)

        # loading...
        time.sleep(1.5)

        # crawling informations on each 7 contents
        # if any selector does not exist, that selector returns blank string
        for i in range(1, 8):
            time.sleep(1)
            try:
                url = driver.find_element_by_css_selector(
                    f'#s_content > div.section > ul > li:nth-child({i}) > dl > dt > a').get_attribute('href') 
            except selenium.common.exceptions.NoSuchElementException:
                url = " "     
            try:
                head = driver.find_element_by_css_selector(
                    f"#s_content > div.section > ul > li:nth-child({i}) > dl > dt > a").text
            except selenium.common.exceptions.NoSuchElementException:
                head = " "
            try:
                contents = driver.find_element_by_css_selector(
                    f"#s_content > div.section > ul > li:nth-child({i}) > dl > dd:nth-child(3)").text 
            except selenium.common.exceptions.NoSuchElementException:
                contents = " "
            try:
                date = driver.find_element_by_css_selector(
                    f"#s_content > div.section > ul > li:nth-child({i}) > dl > dd.txt_inline").text
            except selenium.common.exceptions.NoSuchElementException:
                date = " "

            time.sleep(1)
            print(head)
            print(url)
            print(contents)
            print(date)
            total_header.append(head)
            total_url.append(url)
            total_contents.append(contents)
            total_date.append(date)

    return total_header, total_url, total_contents, total_date

pages = int(input("몇 페이지 까지 크롤링 할까요? \n => "))
keyword = input("검색하실 키워드를 입력해주세요 \n =>")
start_date = input("검색 시작일을 입력해주세요 ex) 2022.01.06. \n =>")
end_date = input("검색 종료일을 입력해주세요 ex) 2022.01.07 \n =>")

# <예시> -> 이렇게 직접 값을 넣어줘야 key error가 안뜸
# keyword = "산업공학과"
# start_date = '2022.01.06.'
# end_date = '2022.01.07' # 여기는 뒤에 점이 없어야 함 
# pages = 1

# print(get_post_info('산업공학과','2022-01-06','2022-01-07',1))

total_header, total_url, total_contents, total_date = get_post_info(keyword, start_date, end_date, pages)
data = {"header": total_header, "url": total_url, "contents": total_contents, "date": total_date}

df = pd.DataFrame(data, columns=["header",'url', "contents", "date"])
keyword = keyword.replace(" ", "_")

# title과 url만 뽑기
kin_title_url = df.iloc[:,:2]
```

# 2. list를 받아와서 블로그 내용 긁기

```python
# Step 1. URL 리스트 불러오기
 
url_load = kin_title_url
num_list = len(url_load)


# Step 2. 리스트에서 하나씩 반복하여 데이터를 수집하기

dict = {}   # 전체 크롤링할 데이터를 담을 그릇 만들기       

# 수집할 글 갯수 정하기 -> 위에서 설정한 것 
number = num_list

# 수집한 URL 반복하여 데이터 수집
for i in range(0, number):      # 0번부터 위에서 지정한 number까지 반복

    # 1_글 띄우기
    url = url_load['url'][i]                                          # 위에서 불러온 url_load에서 url 컬럼의 i번째 값(url)
    driver = webdriver.Chrome("D://chromedriver.exe")            
    driver.get(url)                                                       # 글 띄우기

    # 2_크롤링
    target_info = {} # 개별 블로그별 내용을 담을 딕셔너리 생성
    
        # 질문 제목 
    try :
        overlays = "#content > div.question-content > div > div.c-heading._questionContentsArea.c-heading--default-old > div.c-heading__title > div > div.title"   # F12 개발자도구에서 확인(아래 참고) # div의 class부분을 가져옴 자료 가져옴(내용이 들어있는 부분의 class를 가져와야함)
        title = driver.find_element_by_css_selector(overlays).text     # tlt에 위의 경로에 대한 내용을 가져온다.  
        title                                                                       # text로 가져오는 것을 title로 지정한다.
    except selenium.common.exceptions.NoSuchElementException:
        title = " "

        # 질문 내용
    try :
        overlays = "#content > div.question-content > div > div.c-heading._questionContentsArea.c-heading--default-old > div.c-heading__content" 
        question = driver.find_element_by_css_selector(overlays).text # nickname
    except selenium.common.exceptions.NoSuchElementException:
        question = " "

        # 답변 내용
    try :
        overlays =".se-component-content"
        contents = driver.find_elements_by_css_selector(overlays) # datetime
        
        # contents 역시 위의 tit에서 보여지는 것과 같은 형태로 여러개가 보여진다. 이 때문에 contents에서 텍스트 형태로 내용을 변경해줘야 한다.
        content_list = []  
        for content in contents:                           # contents의 요소 하나씩을 반복하고 이를 content라고 한다.
            content_list.append(content.text)    # content_list안에 content를 텍스트로 가져온 것을 차례대로 가져온다.
        content_str = ' '.join(content_list)         # content_list의 각 문장 사이를 합치고 공백으로 구분한다.
        
    except selenium.common.exceptions.NoSuchElementException:
        content_str = " "

    # 각 요소를 하나의 target_info 딕셔너리에 담기
    target_info['title'] = title
    target_info['question'] = question
    target_info['content'] = content_str

    # 각각의 글을 dict 딕셔너리에 차례로 담기
    dict[i] = target_info
    time.sleep(1)


    # 크롤링 성공 이후, 창 닫기
    # print(i, title) #### 출력
    driver.close()   
        
      # 3_예외 처리

# print('수집한 글 갯수: ', len(dict)) #### 출력
# print(dict)
# [출처] [크롤링] 네이버 블로그 크롤링하기 - 2|작성자 호룡핸리더


# Step 3. 수집된 내용을 파일로 만들어내기

# 판다스로 데이터 프레임 만들기

result_df = pd.DataFrame.from_dict(dict, 'index')   # index를 넣지않으면 컬럼 기준으로 데이터프레임을 만든다.
result_df

```

# 3. 두개의  데이터 프레임 합치기

```python
# 제목/url 합치기 
final_result_df = pd.concat([df,result_df],axis=1) # 처음 데이터와 합치기 
final_result_df.drop(['title'], axis=1,inplace= True) # 결측이 있는 제목 제거하고, 결측이 없는 제목 추출 
final_result_df.drop(['contents'], axis=1,inplace= True) # 요약문 제거 
final_result_df = final_result_df[['url','date','header','question','content']] # 순서 변경 
final_result_df
```

```python
# 날짜 필터링 

target_year_start = '2019'
target_year_end = '2022'

final_result_df = final_result_df[(final_result_df['date'] >= target_year_start+'-01-13') & (final_result_df['date'] <= target_year_end+'-01-13')]
final_result_df

final_result_df.reset_index(drop= False, inplace= True) # reset필요 
final_result_df.drop(['index'],axis=1,inplace= True)
final_result_df
# final_result_df.iloc[:2]
```

```python
# 최종 저장
final_result_df.to_csv("final_kin_content.csv", encoding='utf-8-sig')
```
